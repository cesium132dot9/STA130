### Question 5
We are setting the significance level to $\alpha = 0.05$

```python
import numpy as np
import plotly.graph_objects as go

# Given data
total_couples = 124
right_tilt_observed = 80
probability_right = 0.5
num_simulations = 10000

# Simulating 10,000 trials of flipping a coin 124 times (binomial distribution)
simulated_trials = np.random.binomial(total_couples, probability_right, num_simulations)

# Calculate the p-value: proportion of trials with 80 or more right tilts
p_value = np.mean(simulated_trials >= right_tilt_observed)
print(f"Simulated p-value: {p_value}")

# Create histogram of the simulated trials
hist_data = np.histogram(simulated_trials, bins=20)

# Create figure using plotly
fig = go.Figure()

# Add histogram bars
fig.add_trace(go.Bar(
    x=hist_data[1][:-1],  # bin edges
    y=hist_data[0],       # counts
    marker=dict(color='lightblue'),
    name='Simulated Right Tilts'
))

# Add vertical line for observed right tilts
fig.add_trace(go.Scatter(
    x=[right_tilt_observed, right_tilt_observed],
    y=[0, max(hist_data[0])],
    mode="lines",
    line=dict(color="red", dash="dash"),
    name=f'Observed Right Tilts: {right_tilt_observed}'
))

# Customize layout
fig.update_layout(
    title=f"Distribution of Simulated Right Tilts (p-value: {p_value})",
    xaxis_title="Number of Right Tilts",
    yaxis_title="Frequency",
    showlegend=True
)

# Show figure
fig.show(renderer="png")
```

Using the significance level of $\alpha = 0.5$, the simulated p-value is far smaller than that. Meaning we have very strong evidence against the null hypothesis therefore we can safely reject it.

```python
# Calculate the p-value
# How many bootstrapped statistics generated under H0 
# are "as or more extreme" than the observed statistic 
# (relative to the hypothesized population parameter)? 

observed_statistic = (patient_data.HealthScoreChange>0).mean()
simulated_statistics = IncreaseProportionSimulations_underH0random

# Be careful with "as or more extreme" as it's symmetric!
SimStats_as_or_more_extreme_than_ObsStat = simulated_statistics >= observed_statistic
    
p_value = (SimStats_as_or_more_extreme_than_ObsStat).sum() / number_of_simulations

print("Number of Simulations: ", number_of_simulations, "\n\n",
      "Number of simulated statistics (under HO)\n",
      'that are "as or more extreme" than the observed statistic: ',
      SimStats_as_or_more_extreme_than_ObsStat.sum(), "\n\n",
      'p-value\n(= simulations "as or more extreme" / total simulations): ', p_value, sep="")
```

The test is now one-sided because of the change to "simulated_statistics >= observed_statistic". Instead of checking if the statistics are less than and greater than we are only checking if the simulated statistic is greater than the observed statistic. 
In a one-sided hypothesis test we would expect the p-value to be less than before. Because there are less cases of the simulated sample statistic being as or more extreme than the observed sample statistic.

### Question 8 Fisher's Tea Experiment
The original experiment had a smaller sample size. In our experiment we 10 times the sample size used in the original experiment, meaning that variance of the sample mean for our experiment would be less. 

Null hypothesis ($H_0$): $H_0 \leq 0.5$. The students can't identify which drink was poured first at a rate less than or equal to 50%. 

Alternate hypothesis ($H_A$): $H_A > 0.5$. The students successfully identities which drink was poured first at a success rate above 50%. 

Significance level ($\alpha$): $\alpha = 0.05$ 

```python
import numpy as np

# makes the code reproducible
np.random.seed(42)

# our data
correct_answers = 49
total_students = 80

# proportion of students who correctly identified which drink was poured first 
observed_proportion = correct_answers / total_students

# show the percentage of students who identified correctly 
observed_proportion
```

```python
import plotly.express as px

# generate samples under the assumption that the null hypothesis is true based of our original sample 
n_simulations = 10000
null_hypothesis_proportion = 0.5
simulated_results = np.random.binomial(n=total_students, p=null_hypothesis_proportion, size=n_simulations)

# simulated proportion of students who correctly identified which drink was poured first 
simulated_proportions = simulated_results / total_students

# plot the proportion of the distribution of the simulated sample under the null hypothesis 
fig = px.histogram(simulated_proportions, nbins=30, labels={'value': 'Proportion of Correct Identifications'},
                   title='Distribution of Correct Identifications Under Null Hypothesis (p=0.5)')
# add a vertical line to see the observed proportion of the sample 
fig.add_vline(x=observed_proportion, line_dash="dash", line_color="red", 
              annotation_text=f"Observed ({observed_proportion:.4f})")
# stop the legend from showing on the histogram
fig.update_layout(showlegend=False)

# display the plot
fig.show(renderer="png")
```

```python
# calculating the p-value of a one-sided null hypothesis 
p_value = np.mean(simulated_proportions >= observed_proportion)
p_value
```

Our p-value is 0.0294, while our significance level set ahead of time is 0.05. Therefore we have successfully rejected the null hypothesis. Thus, we also know that the students of STA130 have a greater than 50% chance of correctly identifying which drink was poured first.

Final p-value:
